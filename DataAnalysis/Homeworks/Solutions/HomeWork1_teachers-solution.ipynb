__№1__.  Написать функцию, которая берет на вход прямоугольную матрицу $A$ размера m x n и превращает каждый ее столбец в дискретное распределение с помощью применения функции _softmax_:

$\text{softmax}(a_j) = \frac {\exp (a_j)}{\sum_i \exp(a_i)}$, где a --- столбец матрицы $A$.

Таким образом, в матрице, возвращенной функцией, сумма каждого столбца матрицы должна быть равна 1.

```python
#Пример
```

<font color='red'>Авторское решение:</font> 

```python
def compute_softmax(A):
    expA = np.exp(A)   # вычислили экспоненту
    return expA / expA.sum(axis=0)[np.newaxis, :]  
    # с помощью newaxis оцень удобно нормировать матрицы (о newaxis см. доп. материал на github)
```

<font color='red'>compute_softmax принимает на вход __любую__ прямоугольную матрицу и возвращает ее в преобразованном виде. Проверим нашу функцию на данных из примера:</font> 

```python
A = np.array([[0, 1], [1, 2]])
print compute_softmax(A)
```

<font color='red'>На примере этого задания давайте увидим, что решения с numpy __значительно__ быстрее решений на чистом python. Для этого воспользуемся magic-командой %timeit. Она запускает строчку кода много раз и усредняет время ее выполнения.
Сравним авторское решение с решением через map и через циклы.</font> 

```python
def choose_elems2(X, i, j):
    # решение взято из работы анонимного студента
    arr = []
    np.array(map((lambda x: arr.append(X[i[x], j[x]])), np.arange(len(i))))
    return np.array(arr)
def choose_elems3(X, i, j):
    # наиболее интуитивное решение через циклы
    arr = []
    for ind in range(i.shape[0]):
        arr.append(X[i[ind], j[ind]])
```

__№3__. Написать функцию, которая берет на вход матрицу A размера m x n, вектор b длины m и вектор c длины n и возвращает матрицу, составленную из блоков:

$\begin{pmatrix} A & b \\  c^T& 0 \end{pmatrix}$ = 

$\begin{pmatrix} a_{00} & a_{01} & \dots &a_{0,n-1} & b_0 \\ 
a_{10} & a_{11} & \dots &a_{1,n-1} & b_1 \\
 &&\dots& \\
a_{m-1,0} & a_{m-1,1} & \dots &a_{m-1,n-1} & b_{m-1} \\ 
c_0 & c_1 & \dots & c_{n-1 }& 0 \end{pmatrix}$


```python
#Пример
```

<font color='red'>Авторское решение:</font> 

```python
def construct_matrix(A, b, c):
    # здесь мы очень хотели, чтобы вы пользовались функциями hstack и vstack :)
    # вы ими и пользовались, молодцы :)
    Ab = np.hstack((A, b[:, np.newaxis]))
    c0 = np.hstack((c, [0])) # [0] можно не преобразовывать в массив
    return np.vstack((Ab, c0[np.newaxis, :]))
```

<font color='red'>Тест:</font> 

```python
print construct_matrix(np.array([[0, 1], [1, 2]]), np.array([8, 9]), np.array([7, 3]))
```

__№4__. Даны две небольшие бинарные матрицы $A$ размера $m \times n$ и $B$ размера $n \times k$. 
Написать функцию, возвращающую матрицу $C$ такую, что

$C_{ij} = \&_{s=1}^n (A_{is} \lor B_{sj})$.

Здесь $\&$ означает логическое умножение (and), а $\lor$ — дизъюнкцию (or).

```python
#Пример
```

<font color='red'>Для решения этой задачи очень помогал доп. материал, выложенный на github. Авторское решение:</font> 

```python
def perform_logical_computation(A, B):
    return np.all(np.logical_or(A[:, :, np.newaxis], B[np.newaxis, :, :]), axis=1)
    # функции np.all, np.logical_or подробно расписаны в доп. материале :)
```

<font color='red'>Тест:</font> 

```python
A = np.array([[False, True], [True, False]])
B = np.array([[True, False], [False, True]])
print perform_logical_computation(A, B)
```

<font color='red'>Решение через трехмерный массив полезно для того, чтоб разобраться с логическими функциями и мнимыми осями, но совершенно неэффективно по памяти. Есть другое решение, основанное на двойственности булевых законов. Если поэлементно инвертировать входные матрицы (0->1, 1->0), а затем применить законы де Моргана из дискретной математики, то можно будет воспользоваться обычным умножением матриц. __Это решение приводится для ознакомления, и от вас не требовалось__:</font> 

```python
def perform_logical_computation2(A, B):
    A_ = np.logical_not(A)
    B_ = np.logical_not(B)
    return np.logical_not(A_.dot(B_) > 0)  # > 0 позволяет сопоставить натуральное число бинарному значению.
```

```python
print perform_logical_computation2(A, B)
```

__№5__. Напишите функцию, принимающую на вход np.array размера n и возвращающую матрицу размера n + 1 x n + 1, у которой над главной даигональю стоят элементы массива (остальные элементы должны быть нулевыми). 

```python
#Пример
```

<font color='red'>Авторское решение:</font> 

```python
def make_diagonal(n):
    l = n.shape[0]
    result = np.zeros((l+1, l+1))
    result[np.arange(l), np.arange(l)+1] = n # используем индексацию
    return result
```

```python
# Тестируем:
print make_diagonal(np.array([4, 5, 6, 7, 8]))
```

```python
# Давайте тут снова потестим функцию из №2:
print choose_elems(make_diagonal(np.array([4, 5, 6, 7, 8])), np.arange(5), np.arange(1, 6))
```

__№6__. Дана матрица объекты-признаки X размера n x m. Посчитать (квадратную) матрицу попарных корреляций между признаками C:

$C_{ij} = cor_{ij} = \frac {cov_{ij}}{\sqrt{cov_{ii}cov_{jj}}}$ 

$cov_{ij} = cov(X_i, X_j), \quad i,j = 0, \dots, m-1.$

$cov(\{x_0, \dots, x_{n-1}\}, \{y_0, \dots, y_{n-1}\}) = \frac 1 {n-1} \sum_{i=0} ^ {n-1} (x_i-\bar x) (y_i - \bar y)$.

$\bar x = mean(\{x_0, \dots, x_{n-1}\}) = \frac 1 n \sum_{i=0} ^ {n-1} x_i$

Нельзя использовать функции np.cov, np.corrcoef, np.correlate (их можно использовать для проверки правильности ответа)


<font color='red'>Эту задачу все почему-то испугались, наверно. из-за формул. На самом деле надо сначала посчитать средние, потом матрицу ковариаций, а потом ее просто поделить на дисперсии. Авторское решение:</font>

```python
def compute_corr_matrix(X):
    mean = X.mean(axis=0)[np.newaxis, :]  # на mean ограничения не было, хотя легко и без него считать среднее
    cov = (X-mean).T.dot(X-mean) / (X.shape[0]-1)  # если взглядеться в формулу, то можно увидеть обычное умножение матриц
    return cov / np.sqrt(cov.diagonal()[:, np.newaxis] * cov.diagonal()[np.newaxis, :])   # формула для корреляций
```

```python
X = np.random.rand(100, 4)
print compute_corr_matrix(X)  # наша функция
print np.corrcoef(X.T)   # numpy-функция
print np.all(np.abs(compute_corr_matrix(X)-np.corrcoef(X.T)) < 1e-10)   # сравниваем не ==, а abs() < eps, потому что
                                                                        # числа с плавающей точкой нельзя сравнивать на равенство
```

__№7__. Дан массив X (размерность неизвестна). Заменить все элементы, кратные 3, которые больше 0 и меньше 100, на среднее значение по массиву.

```python
#Пример
```

<font color='red'>Авторское решение:</font> 

```python
def find_some_elements(X):
    X[np.logical_and(np.logical_and(X > 0, X < 100), X % 3 == 0)] = X.mean()
    return X
```

```python
# Проверка:
X = np.random.randint(300, size=20).reshape(4, 5) - 150 # генерируем случайную тестовую матрицу с отр. и пол. числами
print X
print find_some_elements(X)
```

Напишите функцию, которая берет на вход датафрейм pandas со считанными данными и возвращает датафрейм с двумя колонками: буква алфавита и значение ВВП, усреднённое по всем странам, начинающимся с данной буквы.

Для решения задачи может понадобиться функция pandas.groupby, которую вы не успели пройти на семинаре. Поэтому задача и дополнительная.

```python
def compute_means_by_letters(df):   # на самом деле решение почти дано в конспекте семинара
    df2 = df[["Country", "GDP"]]
    return df2.groupby(lambda x: df2.ix[x].Country[0]).aggregate(np.mean)
```

```python
compute_means_by_letters(data)
```
